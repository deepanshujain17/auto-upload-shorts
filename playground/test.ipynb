```python
import os
import pickle
from google_auth_oauthlib.flow import InstalledAppFlow
from google.auth.transport.requests import Request
from googleapiclient.discovery import build

# ---- CONFIGURATION ----
# SCOPES = ["https://www.googleapis.com/auth/youtube.readonly"]
SCOPES = ["https://www.googleapis.com/auth/youtube.force-ssl"]
CLIENT_SECRETS_FILE = "/others/client_secrets.json"
TOKEN_FILE = "/others/token.pkl"

def authenticate_youtube():
    creds = None

    # Load credentials from token file if it exists
    if os.path.exists(TOKEN_FILE):
        with open(TOKEN_FILE, "rb") as token:
            creds = pickle.load(token)

    # If there are no valid credentials, let the user log in
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                CLIENT_SECRETS_FILE, SCOPES)
            creds = flow.run_local_server(port=0)

        # Save credentials
        with open(TOKEN_FILE, "wb") as token:
            pickle.dump(creds, token)

    # Build the YouTube client
    youtube = build("youtube", "v3", credentials=creds)
    return youtube

def get_trending_videos(youtube):
    request = youtube.videos().list(
        part="snippet",
        chart="mostPopular",
        maxResults=5,
        regionCode="IN"
    )
    response = request.execute()

    for item in response.get("items", []):
        snippet = item["snippet"]
        print("📺 Title:", snippet.get("title"))
        print("👤 Channel:", snippet.get("channelTitle"))
        print("📅 Published At:", snippet.get("publishedAt"))
        print("🏷️ Tags:", snippet.get("tags", []))
        print("🔗 Video ID:", item.get("id"))
        print("-" * 60)

if __name__ == "__main__":
    print("🔐 Authenticating...")
    youtube = authenticate_youtube()

    print("📈 Fetching trending videos...")
    get_trending_videos(youtube)
```

```python
# Text to Blank Video Subtitles

from moviepy.video.compositing.CompositeVideoClip import CompositeVideoClip
from moviepy.video.VideoClip import TextClip, ColorClip

def text_to_video(
    text,
    output_filename="output.mp4",
    duration_per_line=3,
    bg_color=(0, 0, 0),
    text_color="white",
    resolution=(1280, 720),
    font_path="/Library/Fonts/Arial.ttf",  # Ensure this path points to a valid .ttf font file on your system
    font_size=48,
):
    lines = [line.strip() for line in text.strip().split("\n") if line.strip()]
    print(lines)
    clips = []

    for line in lines:
        txt_clip = TextClip(
            text=line,
            font=font_path,
            font_size=font_size,
            color=text_color,
        ).with_duration(duration_per_line).with_position("center")

        bg_clip = ColorClip(
            size=resolution,
            color=bg_color,
        ).with_duration(duration_per_line)

        composite = CompositeVideoClip([bg_clip, txt_clip])
        clips.append(composite)

    print(len(clips))
    final_video = CompositeVideoClip(clips)
    final_video.write_videofile(output_filename, fps=24)

# Example usage
sample_text = """This is the first line of subtitle.
Here is the second line.
And this is the third.
"""

text_to_video(sample_text, output_filename="outputs/subtitle_output/subtitled_video.mp4")
```

```python
from moviepy.video.VideoClip import TextClip, ColorClip
from moviepy.video.compositing.CompositeVideoClip import CompositeVideoClip
from moviepy.audio.io.AudioFileClip import AudioFileClip

def text_video_with_audio(
    text,
    audio_path,
    output_path="text_with_audio.mp4",
    resolution=(1280, 720),
    text_color="white",
    bg_color=(0, 0, 0),
    font_path="/Library/Fonts/Arial.ttf",
    font_size=48,
    duration=1,
):
    # Load audio to determine duration
    audio = AudioFileClip(audio_path)

    # Text clip
    txt_clip = TextClip(
        text=text,
        font=font_path,
        font_size=font_size,
        color=text_color,
        size=resolution,
        method='caption',
    ).with_position("center")#.with_duration(duration)

    # Background
    bg_clip = ColorClip(
        size=resolution,
        color=bg_color,
    ).with_duration(duration)

    # Combine text over background
    video = CompositeVideoClip([bg_clip, txt_clip]).with_duration(duration).with_audio(audio)

    # Export
    video.write_videofile(output_path, fps=24)

    audio.close()

    return video
```

```python
from moviepy.video.VideoClip import TextClip, ColorClip, ImageClip
from moviepy.video.compositing.CompositeVideoClip import CompositeVideoClip
from moviepy.audio.io.AudioFileClip import AudioFileClip

def image_audio_to_video(image_path, audio_path, output_path,
                         width, height, fps=24, codec='libx264'):
    """
    Creates a video from a single image and an audio file.

    Parameters:
    - image_path: Path to the input image file.
    - audio_path: Path to the input audio file.
    - output_path: Path where the output video will be saved.
    - width: Desired video width in pixels.
    - height: Desired video height in pixels.
    - fps: Frames per second for the output video (default: 24).
    - codec: Video codec to use (default: libx264).
    """
    # Load audio clip
    audio_clip = AudioFileClip(audio_path)
    duration = audio_clip.duration

    # Create an image clip with the given duration and set size
    image_clip = (ImageClip(image_path)
                  .with_duration(duration)
                  # .resize((width, height))
                  )

    # Attach the audio to the image clip
    video_clip = image_clip.with_audio(audio_clip)

    # Write the final video file
    video_clip.write_videofile(
        output_path,
        fps=fps,
        codec=codec,
        audio_codec='aac'
    )

if __name__ == '__main__':
    # Example usage
    image_path = '../assets/images/bg_image.png'
    audio_path = 'outputs/text_audio/polly_output.mp3'
    output_path = 'outputs/img_audio_video_output/video_output.mp4'
    width = 1280
    height = 720

    image_audio_to_video(
        image_path,
        audio_path,
        output_path,
        width,
        height
    )
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.audio.io.AudioFileClip import AudioFileClip
from moviepy.audio.AudioClip import CompositeAudioClip

def add_audio_to_video_with_background(video_path: str,
                                       audio_path: str,
                                       output_path: str,
                                       bg_volume: float = 0.2,
                                       fg_volume: float = 1.0,
                                       fps: int = 24,
                                       codec: str = 'libx264',
                                       audio_codec: str = 'aac') -> None:
    """
    Adds a foreground audio track to an existing video file while retaining
    the original video audio at a lower background volume.

    Parameters:
    - video_path: Path to the input video file.
    - audio_path: Path to the input audio file.
    - output_path: Path where the output video will be saved.
    - bg_volume: Volume multiplier for the original video audio (default: 0.2).
    - fg_volume: Volume multiplier for the added audio track (default: 1.0).
    - fps: Frames per second for the output video (default: 24).
    - codec: Video codec to use (default: libx264).
    - audio_codec: Audio codec to use (default: aac).
    """
    # Load the original video clip
    video_clip = VideoFileClip(video_path)

    # Extract original audio and adjust its volume via MultiplyVolume
    original_audio = None
    if video_clip.audio:
        original_audio = video_clip.audio.with_volume_scaled(bg_volume)

    # Load the new foreground audio clip and set its volume via MultiplyVolume
    fg_audio = AudioFileClip(audio_path).with_volume_scaled(fg_volume)

    # Combine audio tracks
    tracks = []
    if original_audio:
        tracks.append(original_audio)
    tracks.append(fg_audio)
    combined_audio = CompositeAudioClip(tracks)

    # Attach combined audio to the video clip
    video_with_audio = video_clip.with_audio(combined_audio)

    # Write out the final video file
    video_with_audio.write_videofile(
        output_path,
        fps=fps,
        codec=codec,
        audio_codec=audio_codec
    )

    # Close clips to release resources
    video_clip.close()
    if original_audio:
        original_audio.close()
    fg_audio.close()
    combined_audio.close()
    video_with_audio.close()


if __name__ == '__main__':
    # Example usage
    add_audio_to_video_with_background(
        video_path,
        audio_path,
        output_path,
        bg_volume=0.01,
        fg_volume=1.0
    )
```

```python
def create_video_with_audio(image_path: str,
                            fg_audio_path: str,
                            bg_audio_path: str,
                            output_path: str,
                            fg_volume: float = 1.0,
                            bg_volume: float = 0.2,
                            fps: int = 24):
    # Load the foreground audio and scale its volume
    fg_audio = AudioFileClip(fg_audio_path).with_volume_scaled(fg_volume)
    # Set the duration to match the foreground audio's duration
    duration = fg_audio.duration

    # Load the background audio and scale its volume
    bg_audio = AudioFileClip(bg_audio_path).with_volume_scaled(bg_volume).with_duration(duration)

    # Combine the foreground and background audio tracks
    combined_audio = CompositeAudioClip([fg_audio, bg_audio])

    # Create an image clip with the specified duration and fps
    image_clip = ImageClip(image_path).with_duration(duration).with_fps(fps)

    # Set the combined audio to the image clip
    video = image_clip.with_audio(combined_audio)

    # Write the final video to the output path
    video.write_videofile(output_path, fps=fps, audio_codec='aac')

# Example usage
image_path = '../assets/images/bg_image.png'
bg_audio_path= '../assets/audios/bgm_chubina.mp3'
fg_audio_path= 'outputs/text_audio/polly_output.mp3'
output_path = 'outputs/img_audio_video_output/video_output_img_audio.mp4'

create_video_with_audio(
    image_path,
    fg_audio_path,
    bg_audio_path,
    output_path,
    fg_volume=1.0,
    bg_volume=0.05,
    fps=24
)
```
